# **GroupBy Operations**
1. Group by department and count employees in each department.

2. Find total salary per department.
3. Find average salary per department.
4. Group by city and find max salary.
5. Group by department and gender and count employees.

```python
# Group by department and count employees
print("\n6. Employees count by department:")
employee_df.groupBy("department").count().show()
```

```python
from pyspark.sql.functions import sum

# Total salary per department
print("\n7. Total salary per department:")
employee_df.groupBy("department").agg(sum("salary").alias("total_salary")).show()
```

```python
from pyspark.sql.functions import avg

# Average salary per department
print("\n8. Average salary per department:")
employee_df.groupBy("department").agg(avg("salary").alias("avg_salary")).show()
```

```python
from pyspark.sql.functions import max

# Max salary by city
print("\n9. Max salary by city:")
employee_df.groupBy("city").agg(max("salary").alias("max_salary")).show()
```

```python
# Group by department and gender, count employees
print("\n10. Employee count by department and gender:")
employee_df.groupBy("department", "gender").count().show()
```

# **Multiple Aggregation**
1. Perform multiple aggregations on salary per department:
2. sum
3. average
4. max
5. min
6. Rename all aggregated columns appropriately.
7. Group by department and role, then calculate:
8. total salary
9. average salary
10. employee count
11. Count distinct roles in each department.
12. Find the top 3 departments with the highest total salary.

```python
from pyspark.sql.functions import sum, avg, max, min

# 1. Multiple aggregations on salary per department
print("\n11. Multiple aggregations on salary per department:")
employee_df.groupBy("department").agg(
    sum("salary").alias("total_salary"),
    avg("salary").alias("average_salary"),
    max("salary").alias("max_salary"),
    min("salary").alias("min_salary")
).show()
```

```python
from pyspark.sql.functions import count

# 2. Group by department and role, then calculate total salary, average salary, and employee count
print("\n12. Aggregations by department and role:")
employee_df.groupBy("department", "role").agg(
    sum("salary").alias("total_salary"),
    avg("salary").alias("average_salary"),
    count("employee_id").alias("employee_count")
).show()
```

```python
from pyspark.sql.functions import countDistinct

# 3. Count distinct roles in each department
print("\n13. Distinct roles per department:")
employee_df.groupBy("department").agg(
    countDistinct("role").alias("distinct_roles_count")
).show()
```

```python
# 4. Find the top 3 departments with the highest total salary
print("\n14. Top 3 departments by total salary:")
employee_df.groupBy("department").agg(
    sum("salary").alias("total_salary")
).orderBy("total_salary", ascending=False).limit(3).show()
```

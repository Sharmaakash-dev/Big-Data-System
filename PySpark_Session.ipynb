```python
# Importing pyspark
from pyspark.sql import SparkSession
```

```python
# Start Spark Session
spark = SparkSession.builder \
    .appName("DataFrameBasics") \
    .getOrCreate()
```

```python
# Connect to Drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Read the Dataset
# Example: CSV file from Drive
df = spark.read.option("header", "true").csv("/content/drive/MyDrive/DATASETS/covid_19_data.csv")
df.show()  # Display first 20 rows
```

#**Inspection Methods**

```python
# Or print schema
df.printSchema()
```

```python
# Single column type
dict(df.dtypes)["ObservationDate"]
```

```python
# Check the summary
df.describe().show()  # Count, mean, stddev, min, max
```

#**Column Operations**

```python
# Select column
df.select("Country/Region", "Confirmed").show()

```

```python
# Or using col()
from pyspark.sql.functions import col
df.select(col("Country/Region"), col("Confirmed")).show()
```

```python
#  Adding column
from pyspark.sql.functions import lit
df = df.withColumn("new_coloumn", lit("Deaths"))
```

```python
# Or calculated column
from pyspark.sql.types import DoubleType
df = df.withColumn("total", col("Deaths").cast(DoubleType()) + col("Recovered").cast(DoubleType()))
```

```python
df.show()
```

```python
# Renaming column
df = df.withColumnRenamed("new_coloumn", "total_deaths")
```

```python
df.show()
```

```python
# Dropping column
df = df.drop("total_deaths")
```

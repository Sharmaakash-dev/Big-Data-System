```python
# Start Spark Session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("AdvancedOps").getOrCreate()
```

```python
# Read Dataset
df = spark.read.option("header", "true").csv("/content/drive/MyDrive/DATASETS/covid_19_data.csv")
df.show()
```

#**Filtering Operations**

```python
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType

# Filter Rows by Conditions
# Single condition
df.filter(col("Deaths").cast(DoubleType()) > 25).show()
```

```python
# Multiple columns
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType

df.filter(
    (col("Deaths").cast(DoubleType()) > 25000) &
    (col("Recovered").cast(DoubleType()) < 50000)
).show()
```

#**Logical Operators**

```python
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType

# AND operator (&)
df.filter((col("Deaths").cast(DoubleType()) > 25000) & (df["Country/Region"] == "India")).show()
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType

# OR operator (|)
df.filter((col("Deaths").cast(DoubleType()) > 40000) | (col("Confirmed").cast(DoubleType()) > 80000)).show()
```

```python
# NOT operator (~)
df.filter(~(df["Country/Region"] == "India")).show()
```

#**Comparison Operators**

```python
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType

deaths_col = col("Deaths").cast(DoubleType())

df.filter(deaths_col == 50000).show() # Equal
df.filter(deaths_col != 50000).show() # Not equal
df.filter(deaths_col > 30).show() # Greater than
df.filter(deaths_col >= 30).show() # Greater or equal
df.filter(deaths_col < 30).show() # Less than
df.filter(deaths_col.isin([25, 30, 35])).show() # In list
```

#**Aggregation & Grouping**

##**GroupBy Operations**

```python
# Group by department, aggregate salary
df.groupBy("Country/Region").agg({"Confirmed": "avg"}).show()
# Multiple aggregations
from pyspark.sql.functions import avg, max, count
df.groupBy("Country/Region").agg(
avg("Confirmed").alias("avg_Confirmed"),
max("Confirmed").alias("max_Confirmed"),
count("*").alias("Confirmed_count")
).show()
```

##**Overall Aggregated Values**

```python
from pyspark.sql.functions import sum, avg, min, max, col
from pyspark.sql.types import DoubleType

# Prepare numeric columns by casting to Double
df_numeric = df.withColumn("Deaths", col("Deaths").cast(DoubleType())) \
               .withColumn("Confirmed", col("Confirmed").cast(DoubleType()))

# Single aggregations across entire dataset (removed sum on Country/Region)
df_numeric.agg({"Confirmed": "avg"}).show()

# Using functions with correct casting
df_numeric.agg(
    sum("Deaths").alias("total_Deaths"),
    avg("Confirmed").alias("avg_Confirmed"),
    min("Deaths").alias("min_Deaths")
).show()
```

#**Formatting & Sorting**

##**format_number & Alias**

```python
from pyspark.sql.functions import format_number, round, avg, max, col
from pyspark.sql.types import DoubleType

# Cast Confirmed to DoubleType for numeric operations
df_formatted = df.withColumn("Confirmed", col("Confirmed").cast(DoubleType()))

# Format numbers and show results
df_formatted.groupBy("Country/Region").agg(
    round(avg("Confirmed"), 2).alias("avg_Confirmed"),
    format_number(max("Confirmed"), 2).alias("max_Confirmed_formatted")
).show()
```

#**Sort Data with orderBy**

```python
from pyspark.sql.functions import col, avg

# Single column ascending (using Confirmed instead of salary)
df.orderBy("Confirmed").show()

# Multiple columns (using Country/Region instead of age)
df.orderBy(col("Deaths").desc(), col("Country/Region")).show()

# After groupby (using Country/Region and Confirmed)
df.groupBy("Country/Region").agg(avg(col("Confirmed").cast("double")).alias("avg_Confirmed")) \
.orderBy(col("avg_Confirmed").desc()).show()
```
